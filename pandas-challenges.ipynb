Create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars.Year.value_counts()
#car_year = cars.groupby("Year")
car_year = cars.sort_values("Year")
#print(car_year)
cars_new_index = car_year.set_index("Year")
car_eighty1 =  cars_new_index.loc[1980:1989,]
cars_eighty= car_eighty1.describe().loc["mean":"mean",]
cars_ninety1 = cars_new_index.loc[1990:1999,]
cars_ninety = cars_ninety1.describe().loc["mean":"mean",]

cars_twenty1 = cars_new_index.loc[2000:2009, ]
cars_twenty = cars_twenty1.describe().loc["mean":"mean",]
cars_twenty_ten1 = cars_new_index.loc[2010:2020, ]
cars_twenty_ten =cars_twenty_ten1.describe().loc["mean":"mean",]


cars_1 = pd.concat([cars_eighty,cars_ninety])
cars_2 = pd.concat([cars_1, cars_twenty])
cars_decade1 = pd.concat([cars_2,cars_twenty_ten])
Decade = ["1980", "1990", "2000", "2010"]
cars_decade1["Decade"] = Decade
cars_decade = cars_decade1.set_index("Decade")

cars_decade)
```
