```python
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
```

```python
wiki_url = "https://en.wikipedia.org/wiki/List_of_cities_in_Germany_by_population"
```

```python
response_1 = requests.get(wiki_url)
response_1.status_code
```

```python
#content
soup_5 = BeautifulSoup(response_1.content, 'html.parser')
```

```python
My_table = soup_5.find('table', {'class':'wikitable sortable'})
```

```python
My_table
```

```python
df = pd.read_html(str(My_table))
```

```python
df=pd.concat(df)
```

```python
#mw-content-text > div.mw-parser-output > table.wikitable.sortable.jquery-tablesorter
```

```python
print(df)
```

```python
df.to_csv("Cities&Population_Germany.csv", index= False)
```
